```python
import pandas as pd
from scipy.stats.mstats import winsorize
data=pd.read_csv("../datasets/재무데이터/DB최종재무데이터(0702).csv", encoding="cp949")
```

```python
y=data["부도"]
data.drop("부도", axis=1, inplace=True)
data
```

```python
data.describe()
```

```python
a=data.columns
```

### 윈저라이징 방법(outlier 보정 해주기)

```python
df2=data
```

```python
from scipy import stats
for col in a[3:]:
    df2[col] = stats.mstats.winsorize(df2[col], limits=[0.01, 0.01])
```

```python
df2.describe()
```

```python
df3=pd.concat([df2,y], axis=1)
```

```python
df3.to_csv("../datasets/재무데이터/아웃라이어제거된 최종재무데이터(0703).csv", encoding="cp949", index=False)
```
